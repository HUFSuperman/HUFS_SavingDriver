```python
import re

def extract_data_from_log(log_file):
    with open(log_file, 'r') as file:
        log_data = file.readlines()

    epoch_train_loss = []
    epoch_val_accuracy = []
    temp_train_loss = []
    temp_val_accuracy = []

    for line in log_data:
        # Debug: Print out some lines to understand the log format
        # print(line)

        # Extract training loss for each step
        if "Epoch(train)" in line:
            match = re.search(r"loss: ([\d\.]+)", line)
            if match:
                temp_train_loss.append(float(match.group(1)))
                # Debug: Print when a match is found
                # print(f"Train Loss Match: {match.group(1)}")

        # Calculate average training loss at the end of an epoch
        elif "Training completed for epoch" in line:
            if temp_train_loss:
                average_loss = sum(temp_train_loss) / len(temp_train_loss)
                epoch_train_loss.append(average_loss)
                temp_train_loss = []
                # Debug: Print average training loss
                # print(f"Average Training Loss for Epoch: {average_loss}")

        # Extract validation accuracy for each step
        elif "Epoch(val)" in line:
            match = re.search(r"acc/top1: ([\d\.]+)", line)
            if match:
                temp_val_accuracy.append(float(match.group(1)))
                # Debug: Print when a match is found
                # print(f"Validation Accuracy Match: {match.group(1)}")

        # Calculate average validation accuracy at the end of an epoch
        elif "Validation completed for epoch" in line:
            if temp_val_accuracy:
                average_accuracy = sum(temp_val_accuracy) / len(temp_val_accuracy)
                epoch_val_accuracy.append(average_accuracy)
                temp_val_accuracy = []
                # Debug: Print average validation accuracy
                # print(f"Average Validation Accuracy for Epoch: {average_accuracy}")

    return epoch_train_loss, epoch_val_accuracy

# 아래의 경로로 가서 log파일 경로 가져와서 붙여넣기
train_loss, val_accuracy = extract_data_from_log('./work_dirs/tsn_moblieone_s4_1x1x8_20e_Dassult/20231129_040251/20231129_040251.log')

print(train_loss, val_accuracy)




```

```python
import matplotlib.pyplot as plt

def plot_loss_and_accuracy(train_loss, val_accuracy):
    epochs = range(1, len(train_loss) + 1)

    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Training Loss', color=color)
    ax1.plot(epochs, train_loss, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('Validation Accuracy', color=color)  # we already handled the x-label with ax1
    ax2.plot(epochs, val_accuracy, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()

# Extract data from the log file
train_loss, val_accuracy = extract_data_from_log('./work_dirs/tsn_moblieone_s4_1x1x8_20e_Dassult/20231129_040251/20231129_040251.log')

# Plot the data
plot_loss_and_accuracy(train_loss, val_accuracy)

```
